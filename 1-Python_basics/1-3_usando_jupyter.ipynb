
## Bienvenido a “Comenzando con Jupyter”

Después de ver este video, podrá:

- Describir cómo ejecutar, insertar y eliminar una celda en un cuaderno.
- Trabajar con varios cuadernos.
- Presentar el cuaderno y cerrar la sesión del cuaderno.

---

En la sesión de laboratorio de este módulo, puede iniciar un cuaderno utilizando el entorno virtual de Skills Network. Después de seleccionar la casilla de verificación, haga clic en la pestaña de herramientas **Abrir** y el entorno abrirá Jupyter Lab. Aquí verá el cuaderno abierto.

Al abrir el cuaderno, puede cambiar el nombre del mismo. Haga clic en **Archivo**. Luego, haga clic en **Cambiar nombre del cuaderno** para darle el nombre requerido. Ahora puede comenzar a trabajar en su nuevo cuaderno.

---

### Ejecutar celdas

- En el nuevo cuaderno, escriba “hola mundo”. Luego, haga clic en el botón **Ejecutar** para mostrar que el entorno está dando el resultado correcto.
- En la barra de menú principal en la parte superior, haga clic en **Ejecutar**.
- En el menú desplegable, haga clic en **Ejecutar celdas seleccionadas** para ejecutar las celdas resaltadas actuales. Alternativamente, puede usar un atajo, presionando **Shift + Enter**.
- En caso de que tenga varias celdas de código, haga clic en **Ejecutar todas las celdas** para ejecutar el código en todas las celdas.

---

### Insertar y eliminar celdas

- Puede agregar código insertando una nueva celda. Para agregar una nueva celda, haga clic en el símbolo **+** en la barra de herramientas.
- También puede eliminar una celda. Resalte la celda y, en la barra de menú principal, haga clic en **Editar** y luego en **Eliminar celdas**. Alternativamente, puede usar un atajo presionando **D** dos veces en la celda resaltada.
- Además, puede mover las celdas hacia arriba o hacia abajo según sea necesario.

---

### Trabajar con varios cuadernos

- Para trabajar con varios cuadernos, haga clic en el botón **+** en la barra de herramientas y seleccione el archivo que desea abrir. Se abrirá otro cuaderno.
- Alternativamente, puede hacer clic en **Archivo** en la barra de menú y seleccionar **Abrir un nuevo lanzador** o **Abrir un nuevo cuaderno**. Cuando abra el nuevo archivo, podrá moverlos. Por ejemplo, puede colocar los cuadernos uno al lado del otro.

---

### Comunicación de resultados

- En un cuaderno, puede asignar la variable `uno` al número 1 y la variable `dos` al número 2, y luego imprimir el resultado de sumar los números `uno` y `dos`.
- Como científico de datos, es importante comunicar sus resultados. Jupyter admite la presentación de resultados directamente desde los cuadernos. Puede crear un Markdown para agregar títulos y descripciones de texto para ayudar con el flujo de la presentación.
- Para agregar Markdown, haga clic en **Código** y seleccione **Markdown**.
- Puede crear gráficos de líneas y convertir cada celda y salida en una diapositiva o subdiapositiva en forma de presentación. La funcionalidad de diapositivas en Jupyter le permite entregar código, visualización, texto y salidas del código ejecutado como parte de un proyecto.

---

### Cerrar sesiones de cuaderno

- Ahora, cuando haya terminado de trabajar con su cuaderno o cuadernos, puede apagarlos. Al apagar los cuadernos, se libera su memoria.
- Haga clic en el ícono de parada en la barra lateral, es el segundo ícono desde arriba. Puede finalizar todas las sesiones a la vez o cerrarlas individualmente.
- Después de cerrar la sesión del cuaderno, verá **"sin núcleo"** en la parte superior derecha. Esto confirma que ya no está activo, y ahora puede cerrar las pestañas.
