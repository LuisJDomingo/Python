## Tipos de Datos en Python

Un tipo es cómo Python representa diferentes tipos de datos. En este video, discutiremos algunos tipos ampliamente utilizados en Python.

Puedes tener diferentes tipos en Python, como números enteros, números reales e incluso palabras. Los números enteros, los números reales y las palabras se pueden expresar como diferentes tipos de datos. A continuación se muestra un resumen de tres tipos de datos para los ejemplos dados:

| Expresión | Tipo de Datos |
|-----------|---------------|
| 11        | int           |
| 21.213    | float         |
| "Hola"    | str           |

Podemos ver el tipo de datos real en Python usando el comando `type()`. Los tipos de datos comunes son:

- **`int`**: Representa un número entero.
- **`float`**: Representa un número real, también conocido como flotante.
- **`str`**: Representa una secuencia de caracteres, o una cadena.

---

### Números Enteros

Los números enteros pueden ser negativos o positivos. Aunque hay un rango finito de números enteros, es bastante grande.

### Números Flotantes

Los flotantes incluyen los números enteros, pero también los números decimales. Puedes seleccionar números entre 0 y 1, como 0.5, 0.55, etc. Estos números son flotantes y puedes seguir acercándote a diferentes números con mayor precisión.

---

### Conversión de Tipos

Puedes cambiar el tipo de expresión en Python; esto se llama **encasillamiento**. Por ejemplo:

- Puedes convertir un `int` en un `float`. Por ejemplo, el número entero 2 se convierte en el flotante 2.0.
- Convertir un flotante en un número entero puede resultar en la pérdida de información. Por ejemplo, convertir el flotante 1.1 a entero dará 1.

Si una cadena contiene un valor entero, puedes convertirla a `int`. Pero si la cadena contiene un valor no entero, obtendrás un error.

---

### Conversión a Cadenas

Puedes convertir un `int` en una cadena (`str`) o un `float` en una cadena (`str`).

### Booleanos

Un booleano en Python puede tomar dos valores:

- **`True`** (verdadero, con una T mayúscula)
- **`False`** (falso, con una F mayúscula)

Usando el comando `type()` sobre un valor booleano, obtenemos el término `bool`. Si conviertes un booleano `True` en un número entero o flotante, obtendrás un 1. Si conviertes un booleano `False` a un número entero o flotante, obtendrás un 0.

Si conviertes un 1 a un booleano, obtendrás `True`. De manera similar, si conviertes un 0 a un booleano, obtendrás `False`.

Consulte los laboratorios para obtener más ejemplos o visite [Python.org](https://www.python.org) para conocer otros tipos de datos en Python.
