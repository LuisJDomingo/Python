## Condiciones y Ramificación en Python

En este vídeo, aprenderás sobre las condiciones y la ramificación en Python. Las operaciones de comparación comparan algún valor u operando y producen un valor booleano basado en la condición.

### Operadores de Comparación

- **Igualdad (`==`):**
  - Compara si dos valores son iguales.
  - Ejemplo:
    ```python
    a = 6
    b = 7
    result = (a == b)  # False
    print(result)
    ```

    ```python
    a = 6
    b = 6
    result = (a == b)  # True
    print(result)
    ```

- **Mayor que (`>`):**
  - Compara si el operando izquierdo es mayor que el derecho.
  - Ejemplo:
    ```python
    i = 6
    result = (i > 5)  # True
    print(result)
    ```

- **Mayor o igual que (`>=`):**
  - Compara si el operando izquierdo es mayor o igual que el derecho.
  - Ejemplo:
    ```python
    i = 5
    result = (i >= 5)  # True
    print(result)
    ```

- **Menor que (`<`):**
  - Compara si el operando izquierdo es menor que el derecho.
  - Ejemplo:
    ```python
    i = 2
    result = (i < 6)  # True
    print(result)
    ```

- **Menor o igual que (`<=`):**
  - Compara si el operando izquierdo es menor o igual que el derecho.
  - Ejemplo:
    ```python
    i = 5
    result = (i <= 5)  # True
    print(result)
    ```

- **Desigualdad (`!=`):**
  - Compara si dos operandos no son iguales.
  - Ejemplo:
    ```python
    i = 2
    result = (i != 6)  # True
    print(result)
    ```

### Comparación de Cadenas

- Compara cadenas usando igualdad y desigualdad.
  - Ejemplo:
    ```python
    a = "ACDC"
    b = "Michael Jackson"
    result_eq = (a == b)  # False
    result_neq = (a != b)  # True
    print(result_eq)
    print(result_neq)
    ```

### Ramificación

- **Declaración `if`:**
  - Ejecuta un bloque de código si la condición es verdadera.
  - Ejemplo:
    ```python
    edad = 17
    if edad >= 18:
        print("Entrarás")
    else:
        print("Seguir adelante")
    ```

    ```python
    edad = 19
    if edad >= 18:
        print("Entrarás")
    else:
        print("Seguir adelante")
    ```

- **Declaración `else`:**
  - Ejecuta un bloque de código diferente si la condición `if` es falsa.
  - Ejemplo:
    ```python
    edad = 17
    if edad >= 18:
        print("Entrarás")
    else:
        print("Ir al concierto de Meat Loaf")
    ```

    ```python
    edad = 19
    if edad >= 18:
        print("Entrarás")
    else:
        print("Ir al concierto de Meat Loaf")
    ```

- **Declaración `elif`:**
  - Permite verificar condiciones adicionales si la condición anterior es falsa.
  - Ejemplo:
    ```python
    edad = 18
    if edad >= 19:
        print("Entrarás")
    elif edad == 18:
        print("Ir a ver a Pink Floyd")
    else:
        print("Ir al concierto de Meat Loaf")
    ```

### Operadores Lógicos

- **Operador `not`:**
  - Invierte el valor booleano.
  - Ejemplo:
    ```python
    a = True
    result = not a  # False
    print(result)
    ```

    ```python
    a = False
    result = not a  # True
    print(result)
    ```

- **Operador `or`:**
  - Devuelve `True` si al menos una de las condiciones es verdadera.
  - Ejemplo:
    ```python
    año = 1990
    if año < 1980 or año > 1989:
        print("Este álbum se hizo en los años 70 o 90")
    else:
        print("Este álbum se hizo en los 80")
    ```

- **Operador `and`:**
  - Devuelve `True` solo si todas las condiciones son verdaderas.
  - Ejemplo:
    ```python
    año = 1983
    if año >= 1980 and año <= 1989:
        print("Este álbum se hizo en los 80s")
    else:
        print("Este álbum no se hizo en los 80s")
    ```
