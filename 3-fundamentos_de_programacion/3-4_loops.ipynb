## Bucles en Python

En este vídeo cubriremos los bucles, en particular los bucles `for` y `while`. Usaremos muchos ejemplos visuales en este video. Consulte los laboratorios para ver ejemplos con datos.

### Función `range()`

La función `range()` genera una secuencia ordenada como una lista. 

- **Un parámetro:** Si la entrada es un número entero positivo, la salida es una secuencia que comienza en cero hasta ese número, sin incluirlo.
  - Ejemplo:
    ```python
    # Generar una secuencia de 0 a 4
    seq = list(range(5))  # [0, 1, 2, 3, 4]
    print(seq)
    ```

- **Dos parámetros:** Si la función tiene dos entradas, la secuencia comienza en el primer parámetro y termina antes del segundo parámetro.
  - Ejemplo:
    ```python
    # Generar una secuencia de 10 a 14
    seq = list(range(10, 15))  # [10, 11, 12, 13, 14]
    print(seq)
    ```

### Bucle `for`

Los bucles `for` se utilizan para iterar sobre una secuencia de elementos.

- **Iterar sobre índices:**
  - Ejemplo:
    ```python
    # Lista de colores
    colores = ["rojo", "amarillo", "verde"]
    
    # Reemplazar cada color con "blanco"
    for i in range(len(colores)):
        colores[i] = "blanco"
    
    print(colores)  # ['blanco', 'blanco', 'blanco']
    ```

- **Iterar directamente sobre elementos:**
  - Ejemplo:
    ```python
    # Lista de colores
    colores = ["rojo", "amarillo", "verde"]
    
    # Imprimir cada color
    for color in colores:
        print(color)
    ```

- **Uso de `enumerate()`:**
  - Ejemplo:
    ```python
    # Lista de colores
    colores = ["rojo", "amarillo", "verde"]
    
    # Imprimir índice y color
    for i, color in enumerate(colores):
        print(f"Índice: {i}, Color: {color}")
    ```

### Bucle `while`

Los bucles `while` se ejecutan mientras se cumpla una condición.

- **Ejemplo de bucle `while` para copiar colores:**
  - Ejemplo:
    ```python
    # Lista de colores
    colores = ["naranja", "naranja", "violeta", "naranja"]
    nuevos_colores = []
    
    i = 0
    while i < len(colores) and colores[i] == "naranja":
        nuevos_colores.append(colores[i])
        i += 1
    
    print(nuevos_colores)  # ['naranja', 'naranja']
    ```
